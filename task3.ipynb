EDA

```python

plt.figure(figsize=(10,5))
data_cpy.groupby(['Category'])['Profit'].count().sort_values(ascending = False).plot.bar()
plt.ylabel('Profit')
plt.title('Categories for which profit is highest',fontsize = 20)
plt.show()
```

```python
plt.figure(figsize=(10,5))
data_cpy.groupby(['SubCategory'])['Profit'].count().sort_values(ascending = False).head(10).plot.bar()
plt.ylabel('Profit')
plt.title('Top 10 SubCategories for which profit is highest',fontsize = 20)
plt.show()
```

```python
# sales and subcategory
plt.figure(figsize=(10,5))
data_cpy.groupby(['Category'])['Sales'].count().sort_values(ascending = False).head(10).plot.bar()
plt.ylabel('Sales')
plt.title('Sales on the basis of category',fontsize = 20)
plt.show()
```

```python
# SubCategories for which profit is highest
plt.figure(figsize=(10,5))
data_cpy.groupby(['SubCategory'])['Profit'].count().sort_values(ascending = False).plot.bar()
plt.ylabel('Profit')
plt.title('Profit on the basis of SubCategory',fontsize = 20)
plt.xticks(rotation=50)
plt.show()
```

```python
# SubCategories for which profit is lowest
plt.figure(figsize=(10,5))
data_cpy.groupby(['SubCategory'])['Profit'].count().sort_values(ascending = False).tail(6).plot.bar()
plt.ylabel('Profit')
plt.title('Profit on the basis of SubCategory',fontsize = 20)
plt.xticks(rotation=15)
plt.show()
```

```python
# Sales on different subcategories
plt.figure(figsize=(10,5))
data_cpy.groupby(['SubCategory'])['Sales'].count().sort_values(ascending = False).tail(8).plot.bar()
plt.ylabel('Sales')
plt.title('Sales on the basis of SubCategory',fontsize = 20)
plt.xticks(rotation=15)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(data_cpy.corr(),annot=True)
```

profit and sales have high correlation than the others so as the sales increases profit also increases

```python
# States in which profit is lowest
plt.figure(figsize=(10,5))
data_cpy.groupby(['State'])['Profit'].count().sort_values(ascending = False).tail(18).plot.bar()
plt.ylabel('Profit')
plt.title('Profit v/s States',fontsize = 20)
plt.xticks(rotation=50)
plt.show()
```

```python
# States in which Sales is lowest
plt.figure(figsize=(10,5))
data_cpy.groupby(['State'])['Sales'].count().sort_values(ascending = False).tail(18).plot.bar()
plt.ylabel('Sales')
plt.title('Sales v/s States',fontsize = 20)
plt.xticks(rotation=50)
plt.show()
```

  KEY TAKEAWAYS FROM EXPLORATORY DATA ANALYSIS:

- As the sales for copiers, Machines, Supplies, Fasteners, Bookcases, Envelops, Tables and labels is less as compared to other therefore there is less profit in these products. These are some weak areas where sale and profit is less so we need to work upon these areas to gain more profit.

- There are some states (in the above graph) in which sales and profit both are less, we need to work in these areas to gain more profit.
